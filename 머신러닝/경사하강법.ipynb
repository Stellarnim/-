```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
```

```python
X_train = np.array([10, 22, 30, 38, 50])
Y_train = np.array([41, 45, 62, 75, 85])
```

```python
plt.scatter(X_train, Y_train, color = 'b')
plt.show()
```

```python
np.corrcoef(X_train, Y_train)
```

```python
X_train = X_train.reshape(-1, 1)
print(X_train)
Y_train = Y_train.reshape(-1, 1)
print(Y_train)
```

```python
scalerX = StandardScaler()
scalerX.fit(X_train)
X_train = scalerX.transform(X_train)
print(X_train)
```

```python
scalerY = StandardScaler()
scalerY.fit(Y_train)
Y_train = scalerY.transform(Y_train)
print(Y_train)
```

```python
model = SGDRegressor(verbose=1)
```

```python
model.fit(X_train, Y_train)
```

```python
X_test = np.array([45]).reshape(-1, 1)
X_test = scalerX.transform(X_test)
X_test
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
y_pred_inverse = scalerY.inverse_transform(y_pred)
print(y_pred_inverse)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn import datasets
```

```python
X, y = datasets.load_diabetes(return_X_y = True)
```

```python
X_data = X[:, 9]
print(X_data)
```

```python
g1 = plt.subplot(1, 2, 1)
g1.boxplot(X_data, labels =['s6'])
plt.title('diabetes')
plt.xlabel('Feature')

g2 = plt.subplot(1, 2, 2)
g2.boxplot(y, labels =['disease progression'])
plt.title('diabetes')
plt.xlabel('Target')

plt.show()
```

```python
plt.scatter(X_data, y, color='b')
plt.title('diabetes')
plt.xlabel('s6')
plt.ylabel('disease progression')
plt.show()
```

```python
np.corrcoef(X_data, y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_data, y, test_size=0.3, random_state=1234)
```

```python
x_train = x_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)
```

```python
scalerX = MinMaxScaler()
scalerX.fit(x_train)
x_train = scalerX.transform(x_train)
x_train_scaled = scalerX.transform(x_train)
print(x_train)
```

```python
scalerY = MinMaxScaler()
scalerY.fit(y_train)
y_train = scalerY.transform(y_train)
y_train_scaled = scalerY.transform(y_train)
print(y_train)
```

```python
x_test = x_test.reshape(-1, 1)
x_test_scaled = scalerX.transform(x_test)
y_test = y_test.reshape(-1, 1)
y_test_scaled = scalerY.transform(y_test)
```

```python
model = SGDRegressor(verbose = 1)
```

```python
model.fit(x_train_scaled, y_train_scaled)
```

```python
print('기울기 : ', model.coef_)
print('절편 : ', model.intercept_)
```

```python
plt.scatter(x_train_scaled, y_train_scaled, color = 'b')

x = np.array([min(x_train_scaled), max(x_train_scaled)])
y = model.coef_ * x + model.intercept_
plt.plot(x, y, c = "orange", label = "Regression line")
plt.show()
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
plt.scatter(x_test_scaled, y_test_scaled, color ='g')
plt.scatter(x_test_scaled, y_pred, color = 'b')
plt.show()
```
